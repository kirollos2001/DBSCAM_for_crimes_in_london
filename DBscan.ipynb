```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("miningproject.csv")
df
```

```python
plt.scatter(df.Latitude,df['Longitude'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
```

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df[['Latitude']])
df['Latitude'] = scaler.transform(df[['Latitude']])

scaler.fit(df[['Longitude']])
df['Longitude'] = scaler.transform(df[['Longitude']])
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=6)
labels = dbscan.fit_predict(df[['Latitude','Longitude']])
labels
```

```python
df['cluster'] = labels
df
```

```python
df_1 = df[df.cluster==-1]
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
plt.scatter(df_1.Latitude,df_1['Longitude'],color='pink')
plt.scatter(df1.Latitude,df1['Longitude'],color='green')
plt.scatter(df2.Latitude,df2['Longitude'],color='red')

plt.xlabel('Latitude')
plt.ylabel('Longitude')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_avg = silhouette_score(df[['Longitude', 'Latitude']], dbscan.labels_)
silhouette_avg
```
